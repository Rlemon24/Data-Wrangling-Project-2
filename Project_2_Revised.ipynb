```python


import pandas as pd

datasets = {
    'GDP': '/content/GDP (1).csv',
    'Inflation': '/content/Inflation.csv',
    'Real_GDP': '/content/GDPC1.csv',
    'Unemployment_Rate': '/content/UNRATE.csv',
    'NetExports': '/content/A019RC1A027NBEA.csv',
    'Real_Disposable_Personal_Income': '/content/A067RX1A020NBEA.csv',
    'Per_saving_percentage_disposable_personal_income': '/content/A072RC1A156NBEA.csv',
    'FCTAX': '/content/FCTAX.csv',
    'FYGFD': '/content/FYGFD.csv',
    'GOV_Current_Ex_Welfare_Social': '/content/G160371A027NBEA.csv',
    'Highest_Tax_Rate': '/content/IITTRHB.csv',
    'Real_Median_HouseholdI': '/content/MEHOINUSA672N.csv',
    'Real_Median_PersonalI': '/content/MEPAINUSA672N.csv',
    'FedFundsRate': '/content/RIFSPFFNA.csv',
    'Gov_Total_Exp': '/content/W068RC1A027NBEA.csv',
    'Weekly_Earnings': '/content/LEU0252881600A.csv'
}

dfs = {}
for name, path in datasets.items():
    dfs[name] = pd.read_csv(path)


dfs['FYGFD']['observation_date'] = pd.to_datetime(dfs['FYGFD']['observation_date'])
dfs['FYGFD']['observation_date'] = dfs['FYGFD']['observation_date'].dt.year.apply(lambda x: pd.to_datetime(str(x) + '-01-01'))


for name, df in dfs.items():
    if 'DATE' in df.columns:
        df.rename(columns={'DATE': 'observation_date'}, inplace=True)
    df['observation_date'] = pd.to_datetime(df['observation_date'])


merged_df = dfs['GDP']
for name, df in dfs.items():
    if name != 'GDP':
        merged_df = pd.merge(merged_df, df, on='observation_date', how='outer')



merged_df = merged_df.rename(columns={
    'GDP': 'Gross Domestic Product (Billions)',
    'GDPC1': 'Real Gross Domestic Product - 2017 Dollars (Billions)',
    'UNRATE': 'Unemployment Rate',
    'A019RC1A027NBEA': 'Net Exports (Billions)',
    'A067RX1A020NBEA': 'Real Disposable Personal Income (Thousands)',
    'A072RC1A156NBEA': 'Personal Savings Rate',
    'FCTAX': 'Federal Tax Receipts',
    'FYGFD': 'Federal Government Debt (Billions)',
    'G160371A027NBEA': 'Government Welfare and Social Expenditures',
    'IITTRHB': 'Highest Individual Income Tax Rate',
    'MEHOINUSA672N': 'Real Median Household Income (Thousands)',
    'MEPAINUSA672N': 'Real Median Personal Income (Thousands)',
    'RIFSPFFNA': 'Federal Funds Rate',
    'W068RC1A027NBEA': 'Government Total Expenditures',
    'FPCPITOTLZGUSA':'Inflation',
    'LEU0252881600A':'Weekly Earnings',
    'observation_date':'Date'
})

merged_df
```
